# Project: Fitbit Activity and Sleep Data Analysis

This mini project analyzes the relationship between users’ daily step count and sleep duration using a simulated dataset from smart wearable devices.

**Goal:**  
To identify whether users who get enough sleep (≥ 7 hours) are more physically active compared to others.

**Dataset Columns:**  
- `date`: date  
- `step_count`: number of steps taken daily  
- `sleep_duration`: sleep duration in hours  
- `heart_rate`: average heart rate  

**Tools Used:**  
Python, Pandas, Matplotlib, Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("fitbit_ornek_veri.csv")
df.head()
```

```python
df.describe()
```

```python
print("Average Step Count:", df["adim_sayisi"].mean())
print("Average Sleep Duration:", df["uyku_suresi"].mean())
print("Average Heart Rate:", df["kalp_ritmi"].mean())
```

```python
df.rename(columns={
    "tarih": "date",
    "adim_sayisi": "step_count",
    "uyku_suresi": "sleep_duration",
    "kalp_ritmi": "heart_rate"
}, inplace=True)

df.columns

plt.figure(figsize=(10,5))
sns.lineplot(data=df.sort_values("date"), x="date", y="step_count", color="orange", marker="o")
plt.title("Daily Step Count")
plt.xlabel("Date")
plt.ylabel("Step Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df["sleep_quality"] = df["sleep_duration"] >= 7
df["sleep_quality"].value_counts(normalize=True) * 100
```

•	43% of users get enough sleep (≥ 7 hours)

•	57% of users sleep less than 7 hours, which may negatively impact their physical activity and recovery.


```python
df.groupby("sleep_quality")["step_count"].mean()
```
